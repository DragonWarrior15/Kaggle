```python
# check unique week X center X meal combos
print(len(df.groupby(['week', 'center_id', 'meal_id'])))
```

```python
# add other dataframes to the training df
df = pd.merge(left = df, right = df_center, how = 'left', on = 'center_id')
df = pd.merge(left = df, right = df_meal, how = 'left', on = 'meal_id')
```

```python
# demand as a function of week, work in log of orders
df_orders_week = df.groupby(['week']).apply(lambda x: pd.Series([sum(np.log(x['num_orders'])), 
                                                                 np.std(np.log(x['num_orders']))], 
                                                                index = ['sum', 'dev'])).reset_index()

fig, axs = plt.subplots(nrows = 1, ncols = 1, figsize=(17 , 8))
axs.plot(df_orders_week['week'], df_orders_week['sum'])
axs.plot(df_orders_week['week'], df_orders_week['sum'] + df_orders_week['dev'], ls = '--')
axs.plot(df_orders_week['week'], df_orders_week['sum'] - df_orders_week['dev'], ls = '--')
plt.show()
```

```python
# demand as a function of week, work in log of orders, for different centers
df_orders_week = df.groupby(['week', 'center_id']).apply(lambda x: pd.Series([sum(np.log(x['num_orders'])), 
                                                                 np.std(np.log(x['num_orders']))], 
                                                                index = ['sum', 'dev'])).reset_index()

fig, axs = plt.subplots(nrows = 1, ncols = 1, figsize=(17 , 8))
for center in df_orders_week['center_id'].unique():
    center_filter = df_orders_week['center_id'] == center
    axs.plot(df_orders_week.loc[center_filter, 'week'], df_orders_week.loc[center_filter, 'sum'])
plt.show()
```

```python
# demand as a function of week, work in log of orders, for different centers
df_orders_week = df.groupby(['week', 'cuisine']).apply(lambda x: pd.Series([sum(np.log(x['num_orders'])), 
                                                                 np.std(np.log(x['num_orders']))], 
                                                                index = ['sum', 'dev'])).reset_index()

fig, axs = plt.subplots(nrows = 1, ncols = 1, figsize=(17 , 8))
for center in df_orders_week['cuisine'].unique():
    center_filter = df_orders_week['cuisine'] == center
    axs.plot(df_orders_week.loc[center_filter, 'week'], 
             df_orders_week.loc[center_filter, 'sum'],
             label = center)
axs.legend()
plt.show()
```

```python
# demand as a function of week, work in log of orders, for different centers
df_orders_week = df.groupby(['week', 'region_code']).apply(lambda x: pd.Series([sum(np.log(x['num_orders'])), 
                                                                 np.std(np.log(x['num_orders']))], 
                                                                index = ['sum', 'dev'])).reset_index()

fig, axs = plt.subplots(nrows = 1, ncols = 1, figsize=(17 , 8))
for center in df_orders_week['region_code'].unique():
    center_filter = df_orders_week['region_code'] == center
    axs.plot(df_orders_week.loc[center_filter, 'week'], 
             df_orders_week.loc[center_filter, 'sum'],
             label = center)
axs.legend()
plt.show()
```
