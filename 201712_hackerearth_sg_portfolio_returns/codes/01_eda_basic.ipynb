```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('../inputs/train.csv')
```

```python
col_dsc_dict = {'portfolio_id': 'unique ID',
'desk_id': 'manager ID (the person handling the portfolio, it can be a team)',
'office_id': 'place where portfolio is managed',
'pf_category': 'portfolio category (anonymised)',
'start_date': 'date when portfolio started',
'sold': 'currency (amount sold)',
'euribor_rate': 'euribor lending rate (monthly average)',
'currency': 'currency in which transaction was made',
'libor_rate': 'libor lending rate (monthly average)',
'country_code': 'country where office is',
'bought': 'currency (price at which portfolio is bought)',
'creation_date': 'date when portfolio instruments were traded',
'indicator_code': 'anonymised',
'sell_date': 'date when portfolio is sold',
'type': 'type of portfolio',
'hedge_value': 'if a portfolio got hedged',
'status': 'anonymised',
'return': 'target variable'}
print (('%20s, %5s, %5s, %5s, %7s, %s') % ('column', 'rows', 'uniq', 'nan', 'dtype', 'dsc'))
for col in df.columns.values.tolist():
    try:
        print (('%20s, %5d, %5d, %5d, %7s, %s') % \
           (col, len(df), len(df[col].unique()), df[col].isnull().sum(), df.dtypes[col], col_dsc_dict[col]))
    except KeyError:
        pass
```

```python
df['indicator_code'] = df['indicator_code'].isnull().apply(lambda x: int(x))
hedge_value_dict = {True:1, False:0}
df['hedge_value'] = df['hedge_value'].apply(lambda x: hedge_value_dict[x] if x in hedge_value_dict else 2)
```

```python
df['has_desk_id'] = df['desk_id'].isnull().apply(lambda x: int(x))
df['has_hedge_value'] = df['hedge_value'].apply(lambda x: 0 if x != 2 else 2)
df['status'] = df['status'].apply(lambda x: 1 if x == True else 0)
```

```python
df['bought_sold'] = df['bought'] - df['sold']
df['sell_creation'] = df.apply(lambda row: (datetime.strptime(str(row['sell_date']), '%Y%m%d') - \
                                            datetime.strptime(str(row['creation_date']), '%Y%m%d')).days, axis = 1)
df['sell_year'] = df['sell_date'].apply(lambda x: x//10000)
df['sell_month'] = df['sell_date'].apply(lambda x: (x//100)%100)
df['sell_day'] = df['sell_date'].apply(lambda x: x%100)
```
